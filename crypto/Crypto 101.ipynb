#### The double spending problem
What if Eve pays at the same time 6 coins to Bob, and 6 coins to Alice ? (Eve only has 10 so she can not pay 6 + 6 = 12)<br>
<br>
If a Node1 receives the letter-to-bob first ?

|             | Eve | Bob | Alice |
| ----------- | --- | --- | ----- |
| before      | 10  |  0  |  0    |
|               |     |     |       |
| letter to Bob | -6  | +6  |  -    |
|               |  4  |  6  |  0    |
|               |     |     |       |
| letter to Alice | -6  |  -  |  +6   |
|         | <span style="color:red; font-weight: bold;">-2</span>  |  6  |  6    | <span style="color:red; font-weight: bold;">Illegal</span>


The above node will not honor the letter-to-alice as Eve sent 6 coins to bob and have only 4 coins left, she can't now give 6 coins to alice as she does not have enpugh. So Eve has 4 coins, Bob has 6 coins and Alice has None<br>
However to Node2 which received the letter-to-alice first will only honor the first text it receives, so at the end: Eve has 4 coins, Bob has None and Alice has 4<br>
<br>
In short: Node1 amd Node2 do not agree on who has how much.
<span style="color:Green; font-weight: bold;">So how do we solve this double spend problem ?</span>


#### A notary
Now what if Node10 is a 'notary' ? Node10 would get the two texts from alice in any order, will accept the first and reject the second. This way if everyone goes by what Node10 says, than all will agree on who has how much. ( both Alice and Bob will wait for Node10 to decide if the text they received is valid before sending the merchendies to EVE)<br>
<br>
So Problem solved!<br>

#### An ellected notary and a Block-Chain
Being a notary give huge power that can be abused. With our regular money, it is done by a bank.<br>
However cryptocurrency is a 'Distributed' currency, so a single entity like a bank is not an option<br>
<br>
The solution goes by randomly selecting a notary. each Node can participate in the notary election. One Node is elected as a notary for the next 'Block' - it selects and validates transactions that will fit inside '1 Block' and chain it to the previos Block. (Hence a <b>BlockChain</b>)<br>
Of cource all nodes verify the notary's work (i.e. transaction signatures, no doble spending, no negative balance, etc. ).<br>
<br>
But if the notary doesn't get anything for his job. why should he do it ?

#### The MINERS and comission
When bitcoin was invented, it's inventor Satoshi Nakamoto (pseudo name) issued a limit of 21,000,000 bitcoins. however not all of them were allocated to accounts. he kept some unallocated bitcoins ( hence a 'MINE' ). whenever a notary creates a block, is is awarded some bitcoins (which are removed from the 'MINE'), so a notary is called 'a MINER'.<br>
<br>
On bitcoin a prize of 50 Bitcoins is given to a miner for every block (there is one block every 10 minutes worldwide). The Prize is halfed every 210,000 blocks (or ~ 4 years). When all of the 21,000,000 Bitcoins be mined there will be no prize. however the miners will take a 'comission' for the transactions (they already do that today).Note that The prize today for every block is 12.5 Bitcoins which is about 82,000$ <br>
<br>
How is the comission calculated ? it is not. it is actually a bid - ask process. when Eve wants to send 6 coins to Bob, she actually sends 6 To Bob and whatever amount (e.g. 0.001 coins) to the miner. The miner can fit just so many transactions in a block, so the miner will prefer transactions he earns more from. so if Alice wants her transaction to go through immediatly - she pays more to the miner. if she pays less, her transaction could take more time or not go at all.<br>
How to determine how much comission to pay to the miner ? it actually depends on the global transaction rate and the bitcoin 'wallet' software can ususally find by itself a good comission value.<br>
<br>
There are only so many transactions per block, and only 1 block per 10 minutes, so as there are more transactions - the comission fee goes higher. Note that the comission fee does not depend on the transaction amount (same comission for paying 0.1 bitcoin as for paying 1000 bitcoins), so it is not worthwile to transfer small amounts. <br>

#### Forks
The high comission problem stops bitcoin from being viable for small payments. There could be many solutions to this problem like a higher block rate then once per 10 minutes, or having more transactions per block (i.e. a bigger block size). As bitcoin is distributed and the solution must be unanimous, it is practically impossible to agree on a solution, so a group of people that agree on the solution (e.g. bigger block), change their software, and what we get are two groups that do not agree. The old group continue to maintain 'Bitcoin' and the new group has a different bitcoin, e.g. 'Bitcoin Cash'. note that by both groups agreeing on the history prior to the fork, Everyone who had an amount of bitcoins at the fork time, will have the same amount of the new currency coins.

#### How a miner is elected
As we saw the current compensation is 80000$ per block, not including comission. many wants to be a miner. but only one is needed per 10 minutes. how does bitcoin determine in a non biased way who will be the next miner ?<br>
Well, any miner-wannabe selects transactions to include in the next block as well as a random number for that block, then calculates a Hash (some mathematic algorithm, 'double sha256' in case of bitcoin) on the block (think of it as the sum of all the numbers in the block). if the cacluclated Hash result is small enough, it is the elected miner and can add it's block to the blockchain. if the Hash result is not small enough, the miner-wanabee select a new random number and starts again.<br>
What do we mean by 'small enough' ? the threshold is adjusted as the computing power of the miners changes to keep all miners together finding one such Has per 10 minutes worldwide.
